```python
import matplotlib.pyplot as plt

# Data
hit_data = {
    "Hit Rate (Liked)": {
        "rate": 137 / 243,
        "hits": 137,
        "total": 243
    },
    "Hit Rate (Disliked)": {
        "rate": 118 / 287,
        "hits": 118,
        "total": 287
    },
    "Overall Hit Rate": {
        "rate": (137 + 118) / (243 + 287),
        "hits": 137 + 118,
        "total": 243 + 287
    }
}

# Labels and values
labels = list(hit_data.keys())
values = [v["rate"] for v in hit_data.values()]
annotations = [f'{v["hits"]}/{v["total"]}\n({v["rate"]*100:.0f}%)' for v in hit_data.values()]

colors = ['#81D4FA', '#FF8A80','#4CAF50']

# Plot
plt.figure(figsize=(7, 5))
bars = plt.bar(labels, values, color=colors, edgecolor='gray')

# Annotate each bar
for bar, annotation in zip(bars, annotations):
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        bar.get_height() + 0.005,
        annotation,
        ha='center', va='bottom', fontsize=11
    )

plt.title("Model Hit Rates with Total Tests and Hits", fontsize=14)
plt.ylabel("Hit Rate", fontsize=12)
plt.ylim(0.35, 0.65)  # Zoom in for taller bars
plt.xticks(rotation=15, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Precision value
precision = 0.61

# Data for the plot
metrics_data = {
    "Precision": precision * 100  # Convert to percentage
}

# Color for precision
colors = ['#4CAF50']  # Green for precision

# Plot
plt.figure(figsize=(7, 5))
bars = plt.bar(metrics_data.keys(), metrics_data.values(), color=colors, edgecolor='gray')

# Annotate the bar with the precision value
for bar, value in zip(bars, metrics_data.values()):
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        bar.get_height() + 0.01,
        f'{value:.1f}%',  # Showing value as percentage with 1 decimal place
        ha='center', va='bottom', fontsize=11
    )

plt.title("Precision (in Percent)", fontsize=14)
plt.ylabel("Value (%)", fontsize=12)
plt.ylim(0, 110)  # Extended y-axis for better view of percentage
plt.xticks(rotation=15, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Precision value (for relevant recommendations)
precision = 0.61  # Example precision, replace with actual value
irrelevant = 1 - precision  # Calculate irrelevant recommendations

# Data for the bar chart
metrics_data = {
    "Relevant": precision * 100,  # Convert to percentage
    "Irrelevant": irrelevant * 100  # Convert to percentage
}

# Colors for the bars
colors = ['#4CAF50', '#FF8A80']  # Green for relevant, light red for irrelevant

# Plot the bar chart
plt.figure(figsize=(7, 5))
bars = plt.bar(metrics_data.keys(), metrics_data.values(), color=colors, edgecolor='gray')

# Annotate each bar with the values
for bar, value in zip(bars, metrics_data.values()):
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        bar.get_height() + 0.5,
        f'{value:.1f}%',  # Showing value as percentage with 1 decimal place
        ha='center', va='bottom', fontsize=11
    )

# Title and labels
plt.title("Portion of Relevant vs Irrelevant Recommendations", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=12)
plt.ylim(0, 110)  # Extended y-axis for better view of percentages
plt.xticks(rotation=15, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Labels and scores
labels = [
    "Software Dev & Programming",
    "Data Management & Analytics",
    "Machine Learning & AI",
    "Critical & Creative Thinking",
    "Collaboration & Communication"
]
scores = [3, 3, 3, 2, 2]

# Close the radar plot loop
scores += scores[:1]
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Create radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Plot
ax.plot(angles, scores, color='dodgerblue', linewidth=2)
ax.fill(angles, scores, color='dodgerblue', alpha=0.3)

# X-axis labels (skills)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)

# Y-axis (radial) ticks
ax.set_rgrids([1, 2, 3, 4], angle=0, fontsize=10)

# Optional: set max value
ax.set_ylim(0, 4)

# Title
plt.title("Advisor Evaluation of Skills (4-Point Scale)", fontsize=14, color='navy', pad=20)

plt.tight_layout()
plt.show()
```
